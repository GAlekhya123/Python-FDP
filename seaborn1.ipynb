```python
Matplotlib: Matplotlib is mainly deployed for basic plotting.
Visualization using Matplotlib generally consists of bars, pies, lines, scatter plots and so on.
Seaborn: Seaborn, on the other hand, provides a variety of visualization patterns. 
It uses fewer syntax and has easily interesting default themes.
```

```python
Seaborn builds on top of Matplotlib and introduces additional plot types. 
It also makes your traditional Matplotlib plots look a bit prettier.
```

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sns.set(style='whitegrid')
sinplot()
plt.show()
```

```python
#Removing Axesspines
sns.set_style("white")
sinplot()
sns.despine()
plt.show()
```

```python
sns.set_style("darkgrid", {'axes.axisbelow': False})
sinplot()
sns.despine()
plt.show()
```
