```python
Pie chart
plt.pie()
```

```python
labels = ["Python","Data Analysis","Machine Learning","AI"]
val = [56,67,89,100]
colors = ["gold","magenta","pink","skyblue"]
plt.pie(val,labels=labels,colors = colors,autopct = "%1.2f%%")
plt.axis("equal")
plt.show()
```

```python
labels = ["Python","Data Analysis","Machine Learning","AI"]
val = [56,67,89,100]
colors = ["gold","magenta","pink","skyblue"]
e = (0,2,2,0)
plt.pie(val,labels=labels,colors = colors,autopct = "%1.2f%%",explode=e)
plt.axis("equal")
plt.show()
```

```python
Saving the figure
```

```python

labels = ["Python","Data Analysis","Machine Learning","AI"]
val = [56,67,89,100]
colors = ["gold","magenta","pink","skyblue"]
e = (0,2,2,0)
plt.pie(val,labels=labels,colors = colors,autopct = "%1.2f%%",explode=e)
plt.axis("equal")
plt.savefig("pieplot.jpg") #d://folder name//filename.png
plt.show()
```
