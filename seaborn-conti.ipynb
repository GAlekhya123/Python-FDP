```python
import matplotlib.pyplot as plt
import seaborn as sns
```

- Regression Plot
- lmplot()
- it creates a regression line between 2 parameters and helps to visualize their linear realationship

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips)
```

```python
sns.stripplot(x="total_bill",y="tip",data=tips)
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"])
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="BuPu_r",
           scatter_kws={"s":50})
```

```python
#Displaying multiple plots
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time",hue="smoker")
```

```python
#Displaying multiple plots
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time",hue="smoker",
           height=6,palette="coolwarm")
```

```python
#bin represents how many falls in data
sns.distplot(tips["total_bill"],bins=20)
```
