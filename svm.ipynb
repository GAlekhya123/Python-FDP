```python
import pandas as pd
```

```python
cancer_data = pd.read_csv("cancer.csv")
cancer_data.head()
```

```python
cancer_data.head()
```

```python
cancer_data.columns
```

```python
import seaborn as sns
sns.pairplot(cancer_data,hue="diagnosis",vars=["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean"])
```

```python
input_data = cancer_data.iloc[:,1:31].values
input_data
```

```python
target_data = cancer_data.iloc[:,31].values
target_data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
target_y = le.fit_transform(target_data)
target_y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(input_data,target_y,test_size = 0.3,random_state=1)
```

```python
x_train
```

```python
from sklearn.svm import SVC
```

```python
smn = SVC()
```

```python
smn.fit(x_train,y_train)
```

```python
pred = smn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,pred)
```

```python
c = confusion_matrix(y_test,pred)
```

```python
sns.heatmap(c,annot=True)
```

```python
print(classification_report(y_test,pred))
```
