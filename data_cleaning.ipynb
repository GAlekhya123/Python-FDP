- **Data Imputation**

```python
s = pd.DataFrame({"names":["mounika","alekhya","naga","archana"],"grades":["A",'2',"B",np.nan]})
s
```

```python
s.isna().sum()
```

```python
s.fillna(0)
```

```python
s["grades"]
```

```python
cnt=0
for row in s['grades']:
    try:
        int(row)
        s.loc[cnt, 'grades']=np.nan
    except ValueError:
        pass
    cnt+=1
```
