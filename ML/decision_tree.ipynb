```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
print(iris_data)
```

```python
# Seperate input and output data
input_data = iris_data.data
input_data
```

```python
output_data = iris_data.target
output_data
```

```python

# Seperate the input and output data into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_data,
                                              output_data,
                                              random_state=5,
                                              test_size=0.25)
```

```python

# import required Algorithm
from sklearn.tree import DecisionTreeClassifier
dect = DecisionTreeClassifier()
# max_depth parameter used for mention how many branches 
# we need to print in notebook
```

```python
# train the model using fit method
dect.fit(x_train,y_train)
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(dect,out_file="iris_image.dot",
               feature_names=iris_data.feature_names,
               class_names=iris_data.target_names,
               rounded=True,
               filled = True)
```

```python

dot_data1 = StringIO()
export_graphviz(dect,out_file=dot_data1,
               feature_names=iris_data.feature_names,
               class_names=iris_data.target_names,
               rounded=True,
               filled = True)
pydot_plot = pydotplus.graph_from_dot_data(dot_data1.getvalue())
Image(pydot_plot.create_png())
```
