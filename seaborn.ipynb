```python
x = np.linspace(0, 14, 100)
for i in range(1, 5):
    plt.plot(x, np.sin(x + i * .5) * (7 - i) )
sns.set(style='whitegrid')

plt.show()
```

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sns.set(style='whitegrid')
sinplot()
plt.show()
```

```python
#Removing Axesspines
sns.set_style("white")
sinplot()
sns.despine()
plt.show()
```

```python
sns.set_style("darkgrid", {'axes.axisbelow': False})
sinplot()
sns.despine()
plt.show()
```
