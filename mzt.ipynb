```python
import numpy as np
x = np.linspace(0,14,100)
for i in range(1,5):
    plt.plot(x,np.sin(x+i*0.5)*(7-i))
plt.show()
```

```python
x = np.linspace(0,14,100)
for i in range(1,5):
    plt.plot(x,np.sin(x+i*0.5)*(7-i))
    sns.set_style("darkgrid")
plt.show()
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python

df.shape
```

```python
Count plot¶
```

```python
sns.countplot(x="sex",data=df)
```

```python
df.columns
```

```python
sns.set_style("whitegrid")
sns.countplot(x = "size",data = df)
```

```python
# Scale and context
sns.set_context("poster",font_scale=2) #poster,paper,notebook,talk
plt.figure(figsize=(20,4))
sns.countplot(x="sex",data=df)
```

```python
sns.set_context('paper',font_scale=1)#poster,paper,notebook,talk
plt.figure(figsize=(10,4))
sns.countplot(x='sex',data=df)
```

```python
#stripplot
#stripplot. Draw a scatterplot where one variable is categorical.
#A strip plot can be drawn on its own
```

```python

sns.stripplot(x="total_bill",data=df)
```

```python
sns.set_context("paper",font_scale=2)
sns.stripplot(x="day",y="tip",hue="day",data=df)
```

```python
#Swarm plot
#swarmplot is smilar to strip plot,locations of points are adjusted automatically to avoid 
#the overlapping
```

```python

sns.swarmplot(x="total_bill",data=df)
```

```python
sns.swarmplot(x="day",y="total_bill",data=df)
```

```python
sns.swarmplot(x="tip",y="day",hue="day",data=df)
```

```python

df.columns
```

```python
sns.swarmplot(x="time",y="tip",data=df)
```

```python
sns.swarmplot(x="time",y="tip",data=df,order=["Dinner","Lunch"])
```

```python
#Box plot ot box-and- whisker plot
```

```python
sns.boxplot(x="tip",y="day",data=df)
```

```python
sns.swarmplot(x="tip",y="day",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=df)
```

```python
sns.boxplot(x="tip",y="day",data=df)
sns.swarmplot(x="tip",y="day",data=df,color="0.5") # color= 0 to 1
```

```python
#A violin plot plays a similar role as a box and whisker plot. 
#It shows the distribution of quantitative data across several levels of one 
#(or more) categorical variables such that those distributions can be compared.
```

```python
#Violin plot

sns.violinplot(x="day",y="total_bill",data=df,inner=None)
```

```python
sns.violinplot(x="day",y="total_bill",data=df)
```

```python
sns.boxplot(x="tip",y="day",data=df)
sns.swarmplot(x="tip",y="day",data=df,color="0.5") # color= 0 to 1
sns.stripplot(x="tip",y="day",data=df,color="0.5")
sns.violinplot(x="tip",y="day",data=df)
```

```python
sns.violinplot(x="day",y="total_bill",hue="day",data=df)
```

```python
sns.violinplot(x='day',y='total_bill',data=df,inner=None,hue='size')
```

```python
sns.set_style("darkgrid")
sns.violinplot(x='day',y='total_bill',data=df,hue='sex')
```
